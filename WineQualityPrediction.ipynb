```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

## Data Collection

```python
df = pd.read_csv('winequality-red.csv')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

## Data Analysis and Visualization

```python
#statistical measures of the dataset
df.describe()
```

```python
#number of values for each quality
sns.catplot(x='quality' ,data = df , kind='count')
df['quality'].value_counts()
```

```python
#volatile acidity vs Quality
plot = plt.figure(figsize=(5,5))
sns.barplot(x='quality', y='volatile acidity', data=df)
```

```python
#citric acid vs quality
plot = plt.figure(figsize=(5,5))
sns.barplot(x='quality', y='citric acid', data=df)
```

### Correlation of columns

```python
correlation = df.corr()
```

```python
#construct a heatmap
plt.figure(figsize=(10,10))
plt.title("Heatmap of Correlation")
sns.heatmap(correlation, cbar= True, square=True,fmt='.1f',annot=True, annot_kws ={'size':8},cmap='Blues')
plt.show()
```

```python
#Data Preprocessing
#Seperating Data and Label

x = df.drop(['quality'],axis=1)
print(x)
```

```python
#Label Binarization

y = df['quality'].apply(lambda y_value:1 if y_value>=7 else 0)
print(y)
```

```python
y.value_counts()
```

```python
#Train Test Split

x_train , x_test , y_train, y_test = train_test_split(x,y,test_size=0.2 , random_state= 42)
```

```python
print(x.shape , x_train.shape , x_test.shape)
```

```python
print(y.shape , y_train.shape , y_test.shape)
```

```python
#Random Forest Classifier

model = RandomForestClassifier()
model
```

```python
model.fit(x_train,y_train)
```

```python
#Accuracy Score

x_train_prediction = model.predict(x_train)
train_data_score = accuracy_score(x_train_prediction , y_train)
train_data_score
```

```python
x_test_prediction = model.predict(x_test)
test_data_score = accuracy_score(x_test_prediction , y_test)
test_data_score
```

## Building a Predictive System

```python
## Making a predictive System
input_data=(7.3,0.65,0.0,1.2,0.065,15.0,21.0,0.9946,3.39,0.47,10.0)

#changing the data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

#reshape the np array as the prediction is for single instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)
```

```python
if (prediction[0] == 1):
    print('Good Quality')
else:
    print('Bad Quality')
```
